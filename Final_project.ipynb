# Analysis

## Delivery Analysis

```python
#import vizualization libraries 
from matplotlib import pyplot as plt
import seaborn as sb
```

For business it's always importnant to work properly in every dimension especially when it comes to services they provide to their clients such as Delivery. For example, if client orders a First class delivery than it means that the person wants its delivery as quickly as possible and its importnant for the businees to make it right without any delays. That's why let's firstly look at the distrbibution of number of days that it usually takes to deliver the order after the order was placed for every type of delivery and then we will check the same but for the "First Class" delivery mode..

```python
#Difference = Ship Date - Order Date
df["ShippingDays"] = df["Ship Date"] - df["Order Date"]
df.sort_values(by=['ShippingDays'])
```

```python
#Adding the number of days without the word "days"
df["NumShippingDays"] = df.ShippingDays.apply( lambda x: x.split(' ')[0])
df["NumShippingDays"] = df["NumShippingDays"].astype(int)
df.head(5)
```

```python
#plotting histogram for the number of shipping days
chart = plt.hist(df.NumShippingDays, 
                 bins=7, 
                 color = "#581845",
                )
plt.title('Frequency of the number of shipping days')
plt.xlabel('Number of Shipping Days')
plt.ylabel('Frequency')

plt.show()
```

As it might be seen, the max days it takes for delivery is 7 days, which is very good considering that it also includes dates for Standart delivery. 
Now, let's look at the First class delivery.

```python
#First Class delivery mode
dfFirstClass = df[(df["Ship Mode"] == "First Class")]
dfFirstClass = dfFirstClass.sort_values(["NumShippingDays"], ascending=[False])
dfFirstClass.head(5)
```

```python
#plotting histogram for the number of shipping days for First class
chart = plt.hist(dfFirstClass.NumShippingDays, 
                 bins=5, 
                 color = "#581845",
                )
plt.title('Frequency of the number of shipping days')
plt.xlabel('Number of Shipping Days')
plt.ylabel('Frequency')

plt.show()
```

Here, we can also see that in general it takes 1-3 days for the First Class delivery mode to deliver the order, which is fine. There is only one order which took 5 days to be ordered. Hence, we can conclude that the delivery of this company works perfectly, which for sure satisfy their clients and makes them come back, which is benefitial for the company.

##  Sub-Category Analysis

```python
#Group by category and count the sub-categories
df_cate_subcateg = df.groupby("Category")["Sub-Category"].count()
df_cate_subcateg
```

```python
#Profit ratio grouped by Category
sumProfitPerCateg = df.groupby("Category")["Profit"].sum()
sumSalesPerCateg  = df.groupby("Category")["Sales"].sum()
profitRatioPerCateg = (sumProfitPerCateg /sumSalesPerCateg).reset_index()

#Rename the columns
profitRatioPerCateg.rename(columns={0:'ProfitRatio'}, inplace=True)

profitRatioPerCateg
```

```python
#Add values on bars
def values_on_bars(chart):
    for p in chart.patches:
        chart.annotate(format(p.get_height(), '.3f'), 
                       (p.get_x() + p.get_width() / 2., 
                       p.get_height()), 
                       ha = 'center', 
                       va = 'center',
                       xytext = (0, 10), 
                       textcoords = 'offset points')
```

```python
#Bar chart of the data frame profitRatioPerCateg
plt.figure( figsize = (5, 3) )
chart=sb.barplot(
    data=profitRatioPerCateg, 
    x='Category', 
    y='ProfitRatio',
    palette = "rocket_r",
    ci=None)
chart.set_title('Profit Ratio Per Category')
chart.set_xlabel('Category')
chart.set_ylabel('Profit Ratio')

plt.show()
```

```python
#Calculating sum Profit and sum Sales grouped by Sub-Category
profitRatioPerSubCategory = df.groupby("Sub-Category").agg(sum_profit=('Profit', 'sum'), sum_sales=('Sales', 'sum'))
#Calculating Profit Ratio
profitRatioPerSubCategory["ProfitRatio"] = profitRatioPerSubCategory.sum_profit / profitRatioPerSubCategory.sum_sales

#Sorting by ProfitRatio
profitRatioPerSubCategory = profitRatioPerSubCategory.sort_values(by=['ProfitRatio']).reset_index()
#Selecting first and last row
profitRatioPerSubCategory[ ['Sub-Category','ProfitRatio'] ]
```

```python
#Representing a bar chart of the dataframe profitRatioPerSubCategory with negative(orange) and postive(purple) values
plt.figure( figsize = (15, 12) )
chart=sb.barplot(
    data=profitRatioPerSubCategory, 
    x='Sub-Category',
    y='ProfitRatio', 
    palette = "rocket_r",
    hue = ["Positive" if number >= 0 else "Negative" for number in profitRatioPerSubCategory.ProfitRatio],
    ci=None)
chart.set_title('Profit Ratio per Sub-Category')
values_on_bars(chart)
plt.show()
```

Based on this bart chart we can see the sub-categories that make the company unprofitable are tables , bookcases and supplies, the first two belonging to furtinure and the third one from office supplies

```python
#Calculating sum Profit and sum Sales grouped by Sub-Category With Discount
profitRatioPerSubCategoryDiscount = df.groupby("Sub-Category").agg(sum_profit=('Profit', 'sum'), sum_sales=('Sales', 'sum'), avg_discount = ("Discount","mean"))
#Calculating Profit Ratio
profitRatioPerSubCategoryDiscount["ProfitRatio"] = profitRatioPerSubCategoryDiscount.sum_profit / profitRatioPerSubCategoryDiscount.sum_sales
profitRatioPerSubCategoryDiscount["Discount"] = profitRatioPerSubCategoryDiscount.avg_discount
#Sorting by ProfitRatio
profitRatioPerSubCategoryDiscount = profitRatioPerSubCategoryDiscount.sort_values(by=['ProfitRatio']).reset_index()
#Selecting first and last row
profitRatioPerSubCategoryDiscount[ ['Sub-Category','ProfitRatio', "Discount"] ]
```

```python
#Bar chart representing the Discount per sub-category with the profit ratio of the sub-category
plt.figure( figsize = (19, 7) )
chart=sb.barplot(
    data=profitRatioPerSubCategoryDiscount, 
    x='Sub-Category', 
    y='Discount',
    palette = "rocket_r",
    hue = ["Positif Profit Ratio" if number >= 0 else "Negatif Profit Ratio" for number in profitRatioPerSubCategoryDiscount.ProfitRatio],
    ci=None)
chart.set_title('Discount Per Sub-Category')
chart.set_xlabel('Sub-Category')
chart.set_ylabel('Discount')
values_on_bars(chart)
plt.show()
```

As a next step we wanted to see if discounts is one of the reasons of unprofitability. Hence from this graph we see that the most 
unprofitable sub-categories (tables,bookcases,supplies) have high discounts at the same time , that is one of the reasons 
that reduces company profit.

## Analysis of the Profitability of the States

```python
#Most profitable state
sumProfitPerState= df.groupby("State")["Profit"].sum()
sumSalesPerState  = df.groupby("State")["Sales"].sum()
profitRatioPerState = (sumProfitPerState/sumSalesPerState).reset_index()

#Rename the columns
profitRatioPerState.rename(columns={0:'ProfitRatio'}, inplace=True)

#Sorting the column by the most profitable state
profitRatioPerState = profitRatioPerState.sort_values(by=['ProfitRatio'],ascending=[False]).reset_index(drop= True)

#Top 5 Most Profitable State
MostProfitableState = profitRatioPerState[1:11]
#MostProfitableState

#Top 5 Less Profitable State(Starting from the worst)
LessProfitableState = profitRatioPerState [-10:].sort_values(by=['ProfitRatio'],ascending=[True])
LessProfitableState
```

```python
#Representing a bar chart of the dataframe LessProfitableState and MostProfitableState with negative(orange) and postive(purple) values
plt.figure( figsize = (20, 12) )
plt.subplot(1,2,1)
chart=sb.barplot(
    data=LessProfitableState, 
    x='ProfitRatio',
    y='State', 
    orient = "h",
    color = "#FF5733",
    ci=None)
chart.set_title('Less Profitable State')

plt.subplot(1,2,2)
chart=sb.barplot(
    data=MostProfitableState, 
    x='ProfitRatio',
    y='State', 
    orient = "h",
    color ="#581845",
    ci=None)
chart.set_title('Most Profitable State')

plt.show()
```

We wanted to check the overview of the sales to check which are the most and less profitable states for the companies to aim
new strategies , we can see that Ohio is the lesss profitable state of all of them and Delaware being the most profitable.

## Analysis of Profit Ratio over time

```python
#Order Year and quarter groupped by Profit Ratio 
profitRatioPerYearQuarter = df.groupby(['Order Year','Order Quarter']).agg(sum_profit=('Profit', 'sum'), sum_sales=('Sales', 'sum'))
#Calculating Profit Ratio
profitRatioPerYearQuarter["ProfitRatio"] = profitRatioPerYearQuarter.sum_profit / profitRatioPerYearQuarter.sum_sales

#Sorting by ProfitRatio
profitRatioPerYearQuarter = profitRatioPerYearQuarter.sort_values(by=['ProfitRatio']).reset_index()
#Selecting Order Year, Order Quarter and profit ratio
profitRatioPerYearQuarter[ ['Order Year','Order Quarter','ProfitRatio'] ]
```

```python
#pivot Profit Ratio Per Year and Quarter
pivotProfitRatioPerYearQuarter = profitRatioPerYearQuarter.pivot(
                                columns='Order Year',
                                index='Order Quarter',
                                values='ProfitRatio'
                                )
pivotProfitRatioPerYearQuarter
```

```python
plt.figure( figsize = (13, 5) )
#lineplot representing profitratio per quarter
plt.subplot(1,2,1)
chart=sb.lineplot(data=profitRatioPerYearQuarter, 
                  x='Order Quarter', 
                  y='ProfitRatio',
                  color = "#581845",
                  ci=None)
chart.set_title('Profit ratio per order quarter')
chart.set_xlabel('Quarter')
chart.set_ylabel('Profit Ratio')

#lineplot representing profitratio per year
plt.subplot(1,2,2)
chart=sb.lineplot(data=profitRatioPerYearQuarter, 
                  x='Order Year', 
                  y='ProfitRatio',
                  color = "#581845",
                  ci=None)
chart.set_title('Profit ratio per order year')
chart.set_xlabel('Year')
chart.set_ylabel('Profit Ratio')

plt.show()
```

In the left chart we can se that there was a declive in profit in the third quarter and on the right chart we can see that the
best year are 2016 and 2018.

## Analysis of The Profit Ratio and Discounts with Segments

```python
#Profit ratio grouped by Segment
sumProfitPerSegment = df.groupby(["Segment","Category"])["Profit"].sum()
sumSalesPerSegment  = df.groupby(["Segment","Category"])["Sales"].sum()
profitRatioPerSegment = (sumProfitPerSegment /sumSalesPerSegment).reset_index()

#Rename the columns
profitRatioPerSegment.rename(columns={0:'ProfitRatio'}, inplace=True)

profitRatioPerSegment
```

```python
#Profit Ratio per Order year per category for Corporate Segment
dfCorporate = df[(df["Segment"] == "Corporate")]
dfCorporateProf=dfCorporate.groupby(["Order Year","Category"])["Profit"].sum()
dfCorporateSales  = dfCorporate.groupby(["Order Year","Category"])["Sales"].sum()
dfCorporateRatio = (dfCorporateProf /dfCorporateSales).reset_index()
#Rename the columns
dfCorporateRatio.rename(columns={0:'ProfitRatio'}, inplace=True)

dfCorporateRatio.head(10)
```

```python
#Profit Ratio per Order year per category for Consume Segment
dfConsumer = df[(df["Segment"] == "Consumer")]
dfConsumerProf=dfConsumer.groupby(["Order Year","Category"])["Profit"].sum()
dfConsumerSales  = dfConsumer.groupby(["Order Year","Category"])["Sales"].sum()
dfConsumerRatio = (dfConsumerProf /dfConsumerSales).reset_index()
#Rename the columns
dfConsumerRatio.rename(columns={0:'ProfitRatio'}, inplace=True)

dfConsumerRatio.head(10)
```

```python
#Profit Ratio per Order year per category for Home Office Segment
dfHome = df[(df["Segment"] == "Home Office")]
dfHomeProf=dfHome.groupby(["Order Year","Category"])["Profit"].sum()
dfHomeSales  = dfHome.groupby(["Order Year","Category"])["Sales"].sum()
dfHomeRatio = (dfHomeProf /dfHomeSales).reset_index()
#Rename the columns
dfHomeRatio.rename(columns={0:'ProfitRatio'}, inplace=True)

dfHomeRatio.head(20)
```

```python
#discounts for Corporate
dfDiscCorporate = df[(df["Segment"] == "Corporate")]
dfDiscCorporate=dfDiscCorporate.groupby(["Order Year","Category"])["Discount"].mean().reset_index()
#Rename the columns
dfDiscCorporate.rename(columns={"Discount":"Avg_Discount"}, inplace=True)
dfDiscCorporate.head(10)
```

```python
#Discounts for Consumer
dfDiscConsumer = df[(df["Segment"] == "Consumer")]
dfDiscConsumer=dfDiscConsumer.groupby(["Order Year","Category"])["Discount"].mean().reset_index()
#Rename the columns
dfDiscConsumer.rename(columns={"Discount":"Avg_Discount"}, inplace=True)
dfDiscConsumer.head(20)
```

```python
#Discount for Home office
dfDiscHome = df[(df["Segment"] == "Home Office")]
dfDiscHome=dfDiscHome.groupby(["Order Year","Category"])["Discount"].mean().reset_index()
#Rename the columns
dfDiscHome.rename(columns={"Discount":"Avg_Discount"}, inplace=True)

dfDiscHome.head(20)
```

```python
#lineplot representing Profit Ratio And Disounts per category over the analyzed period  for Segments 

plt.figure( figsize = (27, 15) )

plt.subplot(2,3,1)
chart=sb.lineplot(data =dfCorporateRatio, 
                  x='Order Year', 
                  y='ProfitRatio',
                   palette = "rocket_r",
                  hue="Category",
                  ci=None)
chart.set_title('Profit ratio for Corporate segment per category per year')
chart.set_xlabel('Year')
chart.set_ylabel('Profit Ratio')

plt.subplot(2,3,2)
chart=sb.lineplot(data =dfConsumerRatio, 
                  x='Order Year', 
                  y='ProfitRatio',
                   palette = "rocket_r",
                  hue="Category",
                  ci=None)
chart.set_title('Profit ratio for Consumer segment per category per year')
chart.set_xlabel('Year')
chart.set_ylabel('Profit Ratio')


plt.subplot(2,3,3)
chart=sb.lineplot(data =dfHomeRatio, 
                  x='Order Year', 
                  y='ProfitRatio',
                  palette = "rocket_r",
                  hue="Category",
                  ci=None)
chart.set_title('Profit ratio for Home Office segment per category per year')
chart.set_xlabel('Year')
chart.set_ylabel('Profit Ratio')

#Visualizing Discounts

plt.subplot(2,3,4)
chart=sb.lineplot(data =dfDiscCorporate, 
                  x='Order Year', 
                  y='Avg_Discount',
                  palette = "rocket_r",
                  hue="Category",
                  ci=None)
chart.set_title('Average Discount for Corporate segment per category per year')
chart.set_xlabel('Year')
chart.set_ylabel('Average Discount')

plt.subplot(2,3,5)
chart=sb.lineplot(data =dfDiscConsumer, 
                  x='Order Year', 
                  y='Avg_Discount',
                  palette = "rocket_r",
                  hue="Category",
                  ci=None)
chart.set_title('Average Discount for Consumer segment per category per year')
chart.set_xlabel('Year')
chart.set_ylabel('Average Discount')


plt.subplot(2,3,6)
chart=sb.lineplot(data =dfDiscHome, 
                  x='Order Year', 
                  y='Avg_Discount',
                  palette = "rocket_r",
                  hue="Category",
                  ci=None)
chart.set_title('Average Discount for Home Office segment per category per year')
chart.set_xlabel('Year')
chart.set_ylabel('Average Discount')
```

We decided to analyze if there was a relation between the profit and the quantity of discount per category, and we analyzed it for each segment (Corporate, Consumer, Home Office) separately. 
For the corporate segment we can see that discounts affect the profitability, for technology and office supplies categories we can see that the peeks of profit ratio and discounts coincide, while for furniture in 2017 the company reduced discounts when the profit ration was good, but after that the profit ration gained a downward trend and their attempts to boost profitability applying discount in 2018 didn’t work. In general, we see, that the sales strategy for Technologies for Corporate segment should be changes as the sales drastically went down starting from 2018.

For the Consumer and Home Office segment we can see that their strategy for discounts makes the business unprofitable because whenever they increase discounts for the categories, the Profit ratio drops significantly, and when the discounts are reduced, the profit ration goes up. Hence, the company should look for other ways of boosting sales, because the way they implement their discounts strategy makes the business lose money.
